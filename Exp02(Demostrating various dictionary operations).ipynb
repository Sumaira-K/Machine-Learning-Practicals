```python
student = {
    "Name" : "Sumaira",
    "Age" : 20,
    "Course" : "BTech CSE",
    "Sem" : 4
}
print("Original dictionary: ", student)
```

```python
#Accessing the key value
print("Name: ", student["Name"])
print("Course: ", student.get("Course"))
```

```python
#Adding a new key
student["City"] = "Hoshiarpur"
print("After adding the key: ", student)
```

```python
#Updating the value of a key
student["Sem"] = 5
print("After updating the key: ", student) 
```

```python
#Removing an element
student.pop("Course")
print("After removing: ", student)
student.popitem()
print("After popping: ", student)
```

```python
#Printing keys, values, items
print("keys: ", student.keys())
print("values: ", student.values())
print("items: ", student.items())
```

```python
#Membership
print("Is 'Age' present?", "Age" in student)
```

```python
#Iterating through the dictionary
print("Iterating through the dictionary: ")
for key, value in student.items():
    print(key, ":", value)
```

```python
#Copying the content of dictionary
co = student.copy()
print("Copied dictionary: ",co) 
```

```python
#Clearing the dictionary
co.clear()
print("After clearing: ", co)
```
