```python
#Original list
a = [10,20,30,40,50]
print("Original list: ", a)
```

```python
#Accessing the elements of the list
print("\nFirst element: ", a[0])
print("Last element: ", a[-1])
```

```python
#Adding elements
a.append(70)
print("After appending: ", a)
a.insert(3,88)
print("After inserting: ", a)
```

```python
#Removing elements
a.remove(30)
print("After removing: ", a)
a.pop()
print("After popping: ", a)
```

```python
#Updating the value of any element
a[2] = 77
print("After updating: ", a)
```

```python
#Length of list
print("Length of list: ", len(a))
```

```python
#Sorting the list
a.sort()
print("After sorting: ", a)
```

```python
#rEversing the list
a.reverse()
print("After reversing: ", a)
```

```python
#Membership
print("Is 20 in list?", 20 in a)
```

```python
#Slicing in list
print("Sliced list (index 1 to 3): ",a[1:4]) 
```

```python
#Iterating through the list
print("Iterating through list: ")
for i in a:
    print(i, end = " ")
```
