<h2>Removing Duplicates</h2>

```python
df=df.drop_duplicates()
print(df)
```

<h2>Handling Missing Values</h2>

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
print(df)
```

```python
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
print(df)
```

```python
cab=df["Cabin"].tolist()
print(cab)
```

```python
single_list=[item for element in cab for item in element.split()]
print(single_list)
```

```python
df['Cabin']=np.random.choice(single_list, size=len(df))
print(df)
```

```python
df['Fare']=df['Fare'].fillna(df['Fare'].mean())
print(df)
```

<h2>Verifying Cleaning</h2>

```python
df.isnull().sum()

# Inference:
# No missing values remain in dataset.
```

<h2>Encode Categorical Data</h2>

```python
# Convert gender into numeric
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df
# Inference:
# Categorical data converted into numeric for analysis.
```

<h2>UNIVARIANT(one column)</h2>

```python
"""This graph illustrates the survival count"""
print(df["Survived"].value_counts())
df["Survived"].value_counts().plot(kind="bar")
plt.title("Survival Count")
plt.xticks([0,1],["Not Survived","Survived"])
plt.show()
```

```python
print(df["Sex"].value_counts())
df["Sex"].value_counts().plot(kind="bar")
plt.title("Gender Count")
plt.xticks([0,1],["Male","Female"])
plt.show()

# Inference:
# Male passengers are more than female passengers.
```

```python
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(y=df['Fare'])
plt.title("Fare Boxplot")
plt.show()

# Inference:
# Presence of outliers in fare (some passengers paid very high fares).
```

```python
df['Embarked'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Embarked Distribution")
plt.ylabel('')
plt.show()

# Inference:
# Most passengers boarded from port 'S'.
```

<h2>Bivariate Analysis</h2>

```python
pd.crosstab(df["Sex"], df["Survived"])
```

```python
pd.crosstab(df["Sex"], df["Survived"]).plot(kind="bar")
plt.xticks([0,1],["Male","Female"])
plt.title("Which Gender Survived More?")
plt.show()
```

```python
pd.crosstab(df["Pclass"], df["Survived"])
```

```python
pd.crosstab(df["Pclass"], df["Survived"]).plot(kind="bar")
plt.title("Passenger class survival")
plt.show()
```

```python
plt.scatter(df["Age"], df["Fare"])
plt.title("Age V/S Fare Analysis")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

<h2>Subplots</h2>

```python
plt.figure(figsize=(12,10))

# Survival vs Gender
plt.subplot(2,2,1)
sns.countplot(x='Sex', hue='Survived', data=df)

# Inference:
# Females have higher survival rate.

# Survival vs Class
plt.subplot(2,2,2)
sns.countplot(x='Pclass', hue='Survived', data=df)

# Inference:
# Higher class passengers survived more.

# Age vs Survival
plt.subplot(2,2,3)
sns.boxplot(x='Survived', y='Age', data=df)

# Inference:
# Younger passengers had slightly higher survival.

# Fare vs Survival
plt.subplot(2,2,4)
sns.boxplot(x='Survived', y='Fare', data=df)

# Inference:
# Passengers paying higher fares had better survival chances.

plt.tight_layout()
plt.show()
```

<h2>Numeric Analysis</h2>

```python
corr=df.corr(numeric_only=True)#by default fault
corr
```

```python
plt.imshow(corr, cmap="twilight_shifted")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=60)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

<H2>## Correlation Heatmap
Shows relationship between selected variables.</H2>

```python
corr = df[['Age','Fare','Pclass','Survived']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Inference:
# Fare positively correlates with survival.
# Pclass negatively correlates with survival.
```

<h4>## Pairplot
Shows relationships among multiple variables.</h4>

```python
sns.pairplot(df[['Age','Fare','Pclass','Survived']], hue='Survived')
plt.show()

# Inference:
# Survival is positively related to fare and negatively to class number.
```
