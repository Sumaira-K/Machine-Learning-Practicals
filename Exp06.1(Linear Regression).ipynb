<h4>Import Libraries</h4>

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h4>Create Dataset</h4>

```python
X = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 4, 5, 4, 5, 7])

X = X.reshape(-1, 1)
#Data shows a roughly linear relationship
#As X increases, y also tends to increase
```

<h4>Train-Test Split</h4>

```python
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

<h4>Initialize Parameters</h4>

```python
m = 0   # slope
b = 0   # intercept
lr = 0.01
epochs = 1000

n = len(X_train)
#m represents slope (rate of change)
#b represents intercept (starting value)
#Learning rate controls update speed
```

<h4>Train Model Using Gradient Descent</h4>

```python
for i in range(epochs):
    y_pred = m * X_train.flatten() + b
    
    # gradients
    dm = (-2/n) * np.sum(X_train.flatten() * (y_train - y_pred))
    db = (-2/n) * np.sum(y_train - y_pred)
    
    # update
    m = m - lr * dm
    b = b - lr * db

print("Slope (m):", m)
print("Intercept (b):", b)

#Model learns best-fit line by minimizing error
#Gradient Descent updates m and b iteratively
```

<h4>Strp 5: Predictions</h4>

```python
y_pred_test = m * X_test.flatten() + b

print("Actual:", y_test)
print("Predicted:", y_pred_test)

#Predictions are continuous values
#These values are compared with actual outputs
```

<h4>Step 6: Evaluation Metrics</h4>

```python
# MAE
mae = np.mean(np.abs(y_test - y_pred_test))

# MSE
mse = np.mean((y_test - y_pred_test) ** 2)

# RMSE
rmse = np.sqrt(mse)

# R^2 Score
ss_total = np.sum((y_test - np.mean(y_test)) ** 2)
ss_residual = np.sum((y_test - y_pred_test) ** 2)
r2 = 1 - (ss_residual / ss_total)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)

#MAE shows average error
#MSE penalizes larger errors
#RMSE gives error in same unit as output
#R² shows how well model fits data (closer to 1 is better)
```

<h4>Step 7: Visualization</h4>

```python
plt.scatter(X, y, label='Actual Data')

plt.plot(X, m*X + b, label='Regression Line')

plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
