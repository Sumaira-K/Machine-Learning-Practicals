```python
# ============================================================
# EMPLOYEE DATA GENERATION USING NUMPY
# Fully Structured & Professional Version
# ============================================================

import numpy as np
import pandas as pd


# ============================================================
# FUNCTION TO GENERATE DATA
# ============================================================

def generate_employee_data(n=20, seed=42):
    """
    Generates synthetic employee dataset.

    Parameters:
    n (int): Number of records
    seed (int): Random seed for reproducibility

    Returns:
    pandas.DataFrame
    """

    # ------------------ Validation ------------------
    if n <= 0:
        raise ValueError("Number of records must be greater than 0")

    np.random.seed(seed)

    # ------------------ ID Generation ------------------
    IDs = np.arange(1, n + 1)

    # Introduce 5% duplicate IDs
    duplicate_count = max(1, n // 20)
    duplicate_indices = np.random.choice(n, duplicate_count, replace=False)
    IDs[duplicate_indices] = IDs[duplicate_indices - 1]

    # ------------------ Name Generation ------------------
    names_list = [
        'Kritika', 'Sanjhpreet Kaur', 'Gurmanjit Kaur', 'Sumaira Khan',
        'Ruchika', 'Anshika', 'Sania',
        'Tejpal Singh', 'Sukhmeet Singh', 'Rishabh Banga'
    ]
    Names = np.random.choice(names_list, n)

    # ------------------ Age ------------------
    Age = np.random.randint(21, 60, size=n).astype(float)

    # Introduce 10% missing values
    missing_age = max(1, n // 10)
    age_missing_indices = np.random.choice(n, missing_age, replace=False)
    Age[age_missing_indices] = np.nan

    # ------------------ Gender ------------------
    Gender = np.random.choice(
        ['M', 'F', np.nan],
        size=n,
        p=[0.45, 0.5, 0.05]
    )

    # ------------------ Salary ------------------
    Salary = np.random.randint(25000, 100000, size=n).astype(float)

    # Introduce 10% missing salary
    missing_salary = max(1, n // 10)
    salary_missing_indices = np.random.choice(n, missing_salary, replace=False)
    Salary[salary_missing_indices] = np.nan

    # Introduce 5% duplicate salaries
    duplicate_salary_indices = np.random.choice(n, max(1, n // 20), replace=False)
    Salary[duplicate_salary_indices] = Salary[duplicate_salary_indices - 1]

    # ------------------ Department ------------------
    Department = np.random.choice(
        ['IT', 'Finance', 'Manager', 'Education', np.nan],
        size=n
    )

    # ------------------ City ------------------
    City = np.random.choice(
        ['Nawanshahr', 'Ludhiana', 'Hoshiarpur',
         'Mohali', 'Patiala', 'Amritsar', np.nan],
        size=n
    )

    # ------------------ Experience ------------------
    Experience = np.random.randint(1, 20, size=n)

    # ------------------ Create DataFrame ------------------
    df = pd.DataFrame({
        'ID': IDs,
        'Name': Names,
        'Age': Age,
        'Gender': Gender,
        'Salary': Salary,
        'Department': Department,
        'City': City,
        'Experience': Experience
    })

    return df


# ============================================================
# GENERATE DATA
# ============================================================

df = generate_employee_data(n=50, seed=17)

print("="*60)
print("GENERATED DATASET")
print("="*60)
print(df.head())

print("\nDataset Shape:", df.shape)

# ============================================================
# DATA QUALITY CHECKS
# ============================================================

print("\n" + "="*60)
print("DATA QUALITY SUMMARY")
print("="*60)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate IDs:", df['ID'].duplicated().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nStatistical Summary:")
print(df.describe(include='all'))
```
